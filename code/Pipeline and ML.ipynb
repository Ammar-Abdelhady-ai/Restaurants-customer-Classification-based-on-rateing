### Split Data into X, y
    - X for input
    - y for output

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X.isna().sum()
```

```python
X.shape
```

### using ColumnTransformer to implement change in     categorical columns, 
  - Pipeline to to make all jobs synchronise, 
  - RobustScaler to business outliers

```python
loc_columns = ["location", "listed_in(city)"]


col_trans = ColumnTransformer(
    
    transformers=[
        ("OHE" , OneHotEncoder(sparse=False , drop="first"), ["rest_type"]),
        ("cat", BinaryEncoder(), loc_columns)
    ], remainder="passthrough" 

)
```

### use PCA to dimension reduction

```python
clf = Pipeline(
    steps=[
        ("preprocessor", col_trans),
        ("scaler", RobustScaler()),
#         ("PCA", PCA(0.90))
    ]
)
```

```python
X.shape
```

```python
pipeline = clf.fit(X)
ml_data = pipeline.transform(X)
ml_data.shape
```

```python
joblib.dump(pipeline, r"D:\other\Epsilon\Epsilon Training\project 1\my solution\save files\pipeline.h5")
```

### make **dictionary Machine Learning algorithms and  Functionto** easy write

```python
models = {
    "logisr": LogisticRegression(), 
    "knnclass": KNeighborsClassifier(), 
    "dectree": DecisionTreeClassifier(), 
    "svc": SVC(),
    "rfc": RandomForestClassifier(), 
    "xgbc": XGBClassifier(), 
    "xgbrfc": XGBRFClassifier()
}
```

```python
def fit_model(model, params):
    rs = RandomizedSearchCV(estimator=model, param_distributions=params, cv=10, 
                              scoring='accuracy', n_jobs=-1, verbose=3)
    rs.fit(ml_data, y)
    return rs
```

### **hyperparameter tuning**

```python
rfc_params = [
    {'criterion': ['gini', 'entropy'], 'max_depth': range(1, 40), 'n_estimators': range(100, 201, 20)}
]

rfc_rs = fit_model(models["rfc"], rfc_params)
rfc_rs.best_score_
```

```python
joblib.dump(rfc_rs, r"D:\other\Epsilon\Epsilon Training\project 1\my solution\save files\model.h5")
```

```python
logisr_params = [
    {'penalty': ['l1', 'l2', 'elasticnet'], 'C': [1, 10, 100, 1000]}
]

logisr_rs = fit_model(models["logisr"], logisr_params)
logisr_rs.best_score_
```

```python
dectree_params = [
    {'criterion': ['gini', 'entropy'], 'max_depth': range(1, 100)}
]

dectree_rs = fit_model(models["dectree"], dectree_params)
dectree_rs.best_score_
```

```python
knnclass_params = [
    {'n_neighbors': range(5, 200, 2)}
]

knnclass_rs = fit_model(models["knnclass"], knnclass_params)
knnclass_rs.best_score_
```

```python
xgbc_params = [
    {'criterion': ['gini', 'entropy'], 'max_depth': range(1, 20), 'n_estimators': range(100, 501, 50)}
]

xgbc_rs = fit_model(models["xgbc"], xgbc_params)
xgbc_rs.best_score_
```

```python
svc_params = [
    {'C': [1, 10, 100, 1000], 'kernel': ['rbf'], 'gamma': [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]},
    {'C': [1, 10, 100, 1000], 'kernel': ['poly'], 'degree': range(2, 20)},
    {'C': [1, 10, 100, 1000], 'kernel': ['sigmoid']}
]

svc_rs = fit_model(models["svc"], svc_params)
svc_rs.best_score_
```
