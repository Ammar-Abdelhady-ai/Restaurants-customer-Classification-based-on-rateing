### Fix rate colum to extract target column

```python
def fix_rate(row):
    try:
        return float(row[:-2])
    except:
        return row
```

```python
df["rate"] = df["rate"].apply(fix_rate)
df["rate"]
```

```python
df["rate"].unique()
```

### threshold to rate: 
####  if  rate >= 3.5 
    - then label : 1
    
    
#### else 
    - label : 0

```python
def git_target(row):
    try:
        if row >= 3.5:
            return 1
        return 0
    
    except:
        return 0
```

```python
df["target"] = df["rate"].apply(git_target)
df["target"]
```

### make DataFrame to show categorical column info :

```python
column_name = []
n_unique = []
n_nan = []

for cat in cats:
    column_name.append(cat)
    n_unique.append(df[cat].nunique())
    n_nan.append(df[cat].isna().sum())
    
    
data = {"column name": column_name, "nan number": n_nan, "unique number": n_unique}
cat_info = pd.DataFrame(data)
cat_info
```
